## 2014 Chicago Divvy Data

## Weather - Average High Temp by Month (F) 

```python
# create dictionaries for average monthly high for each year (2014-2017)
wx_2014_df = pd.DataFrame({
    "Month":["January","February","March","April","May", "June", "July",
            "August", "September", "October", "November", "December"],
    "Temperature":[25,26,41,59,71,81,80,82,74,61,41,37]})
wx_2014_df['Year'] = 2014


wx_2015_df = pd.DataFrame({
    "Month":["January","February","March","April","May", "June", "July",
            "August", "September", "October", "November", "December"],
    "Temperature":[29,23,44,60,70,77,82,81,78,63,54,44]})
wx_2015_df['Year'] = 2015

wx_2016_df = pd.DataFrame({
    "Month":["January","February","March","April","May", "June", "July",
            "August", "September", "October", "November", "December"],
    "Temperature":[32,37,51,57,70,82,84,84,78,66,55,31]})
wx_2016_df['Year'] = 2016

wx_2017_df = pd.DataFrame({
    "Month":["January","February","March","April","May", "June", "July",
            "August", "September", "October", "November", "December"],
    "Temperature":[35,48,47,64,68,83,81,81,80,65,47,34]})
wx_2017_df['Year'] = 2017

wx_avg_df = pd.DataFrame({
    "Month":["January","February","March","April","May", "June", "July",
            "August", "September", "October", "November", "December"],
    "Temperature":[30.25,33.5,45.75,60,69.75,80.75,81.75,82,77.5,63.75,49.25,36.5]})
```

```python
# plot the weather for each year
yearly_wx = pd.concat([wx_2014_df, wx_2015_df, wx_2016_df, wx_2017_df])

colors = ["#6091c3", "#1f2f57", "#a5acaf", "#66c010"]

sns.pointplot(x='Month', y='Temperature', hue='Year', data=yearly_wx, palette = colors)
plt.title("Yearly Weather (Avg High in Fehrenheit)", fontsize=20)
plt.ylabel("Temperature(F*)", fontsize=16)
plt.xlabel("Months", fontsize=16)
#plt.xticks(rotation=90)
plt.grid(linestyle='dotted')
plt.tight_layout()
plt.rcParams["figure.figsize"] = [16,9]
plt.legend(fontsize=13)
plt.savefig("Agregate Average Weather Highs.png")

plt.show()
```

## Identify How Many Stations & Bikes Exist Each Year

```python
# Count unique stations using from_station_id
station_count_2014 = divvy_2014['from_station_id'].nunique()
station_count_2015 = divvy_2015['from_station_id'].nunique()
station_count_2016 = divvy_2016['from_station_id'].nunique()
station_count_2017 = divvy_2017['from_station_id'].nunique()

# Count unique bikes in operation
bike_count_2014 = divvy_2014['bikeid'].nunique()
bike_count_2015 = divvy_2015['bikeid'].nunique()
bike_count_2016 = divvy_2016['bikeid'].nunique()
bike_count_2017 = divvy_2017['bikeid'].nunique()

station_bike_count = pd.DataFrame({
    "Year":[2014,2015,2016,2017],
    "Stations":[station_count_2014,station_count_2015,station_count_2016,station_count_2017],
    "Bikes":[bike_count_2014,bike_count_2015,bike_count_2016,bike_count_2017]
})

station_bike_count = station_bike_count[["Year", "Stations", "Bikes"]]
```

```python
# plot out multiple figures showing growth in Divvy stations and bikes
plt.figure(1)
plt.subplot(211)
sns.pointplot(x='Year', y='Stations', data=station_bike_count, color='#6091c3')
title = plt.title("Growth in Divvy Stations and Bikes", fontsize=20)
title.set_position([.5, 1.05])
bbox_props = dict(boxstyle="round,pad=0.3", fc="w", ec="black", lw=1)
t = plt.text(3, 550, "95% Growth in # of Stations", ha="center", va="center", rotation=00, size=13, bbox=bbox_props)
plt.tick_params(axis='x', bottom='off', labelbottom='off')
plt.ylabel("Station Count", fontsize=16)
plt.xlabel("Year", fontsize=0)
plt.grid(linestyle='dotted')
plt.tight_layout()

plt.subplot(212)
sns.pointplot(x='Year', y='Bikes', data=station_bike_count, color='#1f2f57')
bbox_props = dict(boxstyle="round,pad=0.3", fc="w", ec="black", lw=1)
t = plt.text(3, 5700, "110% Growth in # of Bikes", ha="center", va="center", rotation=00, size=13, bbox=bbox_props)
plt.ylabel("Bike Count", fontsize=16)
plt.xlabel("Year", fontsize=16)
plt.grid(linestyle='dotted')
plt.tight_layout()

plt.rcParams["figure.figsize"] = [16,9]
plt.legend(fontsize=13)
plt.savefig("Yearly Bike and Station Growth.png")

plt.show()
```
